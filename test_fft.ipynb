```python
import torch
import numpy as np
import matplotlib.pyplot as plt

# 假设信号长度
T = 128
# 采样频率，这里假设每周期采样点数为T
fs = 1
# 频率，比如1Hz的正弦波
f = 1
# 时间轴
t = torch.linspace(0, T/fs, T, dtype=torch.float32)
# 构建正弦波信号
x_time_domain = torch.sin(2 * np.pi * f * t)

# 绘制时域信号
plt.figure(figsize=(10, 5))
plt.plot(t.numpy(), x_time_domain.numpy())
plt.title('Time Domain Signal')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.show()
```

```python
# 使用rfft进行变换
xf = torch.fft.rfft(x_time_domain, dim=0)

# 计算对应的频率轴（对于rfft，由于是对实数信号的变换，频率轴的范围是从0到fs/2）
n = xf.size(-1)  # 由于进行了rfft，长度为原长度一半+1
frequencies = torch.linspace(start=int(0), end=int(fs / 2), steps=n, endpoint=True)
```

```python
# 绘制频域幅度谱
plt.figure(figsize=(10, 5))
plt.plot(frequencies.numpy(), torch.abs(xf).numpy())
plt.title('Frequency Domain Amplitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.show()
```
